# FlyThruGateCurriculumAvitary smoke tests

Chạy nhanh để kiểm tra curriculum env:
- Tạo env, in observation/action space.
- Reset, quan sát info và kích thước obs.
- Chạy vài bước hành động ngẫu nhiên.
- Xem từng level spawn với GUI bật để quan sát trực quan.

```python
from gym_pybullet_drones.envs.FlyThruGateCurriculumAvitary import FlyThruGateCurriculumAvitary
from gym_pybullet_drones.utils.enums import ObservationType, ActionType
import numpy as np

# Tạo env không GUI để chạy nhanh
env = FlyThruGateCurriculumAvitary(
    obs=ObservationType('kin'),
    act=ActionType('rpm'),
    gui=False,
    record=False,
    start_level=0,
)

print("Observation space:", env.observation_space)
print("Action space:", env.action_space)
```

```python
# Reset và quan sát info ban đầu
obs, info = env.reset()
print("Initial obs shape:", np.array(obs).shape)
print("Initial info:", info)
```

```python
# Chạy vài bước với hành động ngẫu nhiên
for step in range(5):
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    print(
        f"Step {step}: reward={reward:.3f}, terminated={terminated}, truncated={truncated}, level={info.get('curriculum_level')}, spawn_range={info.get('spawn_range')}"
    )

env.close()
```

## Quan sát spawn từng level với GUI bật

Tạo env mới cho từng level, bật GUI để xem vị trí spawn và chạy vài bước với action = 0.

```python
from time import sleep

for lvl in range(5):  # điều chỉnh nếu num_levels khác 5
    print(f"\n=== Level {lvl} ===")
    sim_env = FlyThruGateCurriculumAvitary(
        obs=ObservationType('kin'),
        act=ActionType('rpm'),
        gui=True,
        record=False,
        start_level=lvl,
    )
    obs, info = sim_env.reset()
    print("spawn_range:", info.get("spawn_range"), "| position:", obs[0][:3])

    zero_action = np.zeros_like(sim_env.action_space.sample())
    for _ in range(10):
        obs, reward, terminated, truncated, info = sim_env.step(zero_action)
        sleep(0.05)
        if terminated or truncated:
            break
    sim_env.close()
```
