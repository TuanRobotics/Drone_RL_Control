```python
import time, pybullet as p, pkg_resources
cid = p.connect(p.GUI)
p.setGravity(0,0,-9.8)
gate = p.loadURDF(pkg_resources.resource_filename('gym_pybullet_drones','assets/gate.urdf'),
                  [0, -1.0, 0.0], p.getQuaternionFromEuler([1.0, 1.0, 1.57]))
drone = p.loadURDF(pkg_resources.resource_filename('gym_pybullet_drones','assets/cf2x.urdf'),
                   [0, 1.0, 0.1])
for i in range(240*5):
    p.stepSimulation()
    if p.getContactPoints(drone, gate):
        print("CONTACT at step", i)
        break
    time.sleep(1/240)
p.disconnect()
```
